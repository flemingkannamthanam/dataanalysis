Now we have to check for any missing data.

```python
Nancheck = raindata.isnull()
Nancheck.sum()
```

Now we find that NaN values nexist in our data frame. The column ANUAL have maximum mumber of missing values. It have 26 NaN values. In case of rainfall data, we can't take average of values to replace a missing data. So better to eliminate missing rows

```python
columns = list(raindata.columns)
columns
```

```python
raindata1 = raindata.drop(raindata[Nancheck.ANNUAL == True].index)
raindata1
```

```python
raindata1.isnull().sum()
```

Now we have removed NaN values completely.

```python
raindata1.shape
```

We have 4090 rows and 19 columns now with no values missing. By analysing the data there are chances of error in recorded values. It seems to gave 0 rainfall in some places even in month of July and August. But the numbers are very less and we can neglect them

```python
rainless = raindata[raindata.AUG < 1]
rainless
```

```python
raindatastate = raindata1.groupby('SUBDIVISION').ANNUAL.sum()
raindatastate
```

Let us make a histogram to see the variation


```python
plt.figure(figsize = (25,15))
plt.xticks(rotation = 90)
plt.bar(raindatastate.index, height = raindatastate)
```

Now we can easily identify Coastal Karnataka region have highest rainfall over 1901 to 2015. Also West Rajasthan have lowest rainfall Now let us go for maximum rainfall of a region


```python
raindatastate1 = raindata1.groupby('SUBDIVISION').ANNUAL.max()
raindatastate1
```

```python
plt.figure(figsize = (25,15))
plt.xticks(rotation = 90)
plt.bar(raindatastate1.index, height = raindatastate1)
```

Now the maximum values of annual rain  fall is recorded in Arunachal Pradesh followed by Coastal Karnataka and minimum is recorded in West Rajasthan. So let us have a more detailed study of following regions
